```python
# Function to remove emojis from text using the 'emoji' library
def remove_emojis(text):
    return emoji.replace_emoji(text, replace='')

# Alternative function to remove emojis using regex (works without the 'emoji' library)
def remove_emojis(text):
    emoji_pattern = re.compile(
        "["
        "\U0001F600-\U0001F64F"  # Emoticons
        "\U0001F300-\U0001F5FF"  # Symbols & pictographs
        "\U0001F680-\U0001F6FF"  # Transport & map symbols
        "\U0001F1E0-\U0001F1FF"  # Flags (iOS)
        "\U00002702-\U000027B0"
        "\U000024C2-\U0001F251"
        "]+", flags=re.UNICODE)
    return emoji_pattern.sub(r'', text)

# Function for general text preprocessing
def preprocess_text(text):
    # 1. Convert text to lowercase
    text = text.lower()

    # 2. Remove URLs from the text
    text = re.sub(r'http\S+|www\S+|https\S+', '', text, flags=re.MULTILINE)

    # 3. Remove emojis from the text
    text = remove_emojis(text)

    # 4. Remove punctuation and special characters, keep only letters and spaces
    text = re.sub(r'[^a-z\s]', '', text)

    # 5. Tokenize the text into words
    tokens = word_tokenize(text)
# 6. Remove English stopwords from the tokens
    stop_words = set(stopwords.words('english'))
    filtered_tokens = [w for w in tokens if w not in stop_words]

    # 7. Join tokens back into a single cleaned string
    cleaned_text = " ".join(filtered_tokens)

    return cleaned_text
```

# Part 8 - Sentiment Analysis with Vader

```python
# Initialize the VADER sentiment analyzer
sia = SentimentIntensityAnalyzer()

# Function to classify sentiment based on VADER compound score
def vader_sentiment_classify(text):
    scores = sia.polarity_scores(text)
    compound = scores['compound']
    if compound >= 0.05:
        return 'positive'
    elif compound <= -0.05:
        return 'negative'
    else:
        return 'neutral'

# Calculate and add the compound sentiment score for each cleaned review
df['compound_score'] = df['cleaned_v2'].astype(str).apply(lambda x: sia.polarity_scores(x)['compound'])

# Apply the function on the cleaned text (or original text if preferred)
df['sentiment'] = df['cleaned_v2'].astype(str).apply(vader_sentiment_classify)

# Display the first 10 rows of cleaned reviews and their corresponding VADER sentiment labels
print(df[['cleaned_v2', 'sentiment']].head(n=25))
```

```python
def plot_unique_top_tokens(df, target_sentiment, other_sentiments, top_n=20, ax=None, palette='Blues'):

    # All tokens for the target sentiment
    target_tokens = [token for sublist in df[df['sentiment'] == target_sentiment]['tokens'] for token in sublist]

    # All tokens for the other sentiment(s)
    other_tokens = set(token for sublist in df[df['sentiment'].isin(other_sentiments)]['tokens'] for token in sublist)

# Keep only tokens that are not in other sentiments
    unique_tokens = [token for token in target_tokens if token not in other_tokens]
    token_counts = Counter(unique_tokens)
    most_common = token_counts.most_common(top_n)
    if most_common:
        words, counts = zip(*most_common)
        sns.barplot(x=list(words), y=list(counts), ax=ax, palette=palette)
        ax.set_title(f'Top {top_n} Unique Tokens - {target_sentiment.capitalize()}')
        ax.set_xlabel('Tokens')
        ax.set_ylabel('Frequency')
        ax.tick_params(axis='x', rotation=45)
        ax.grid(axis='y', linestyle='--', alpha=0.7)

        # Annotate bars
        for p in ax.patches:
            height = p.get_height()
            ax.annotate(f'{height}', (p.get_x() + p.get_width() / 2, height),
                        ha='center', va='bottom', fontsize=10)
    else:
        ax.set_title(f'No unique tokens for {target_sentiment}')

# Exemplo de uso:
fig, axes = plt.subplots(1, 2, figsize=(20, 10))
plot_unique_top_tokens(df, 'negative', ['neutral'], ax=axes[0], palette='Reds')
plot_unique_top_tokens(df, 'neutral', ['negative'], ax=axes[1], palette='Blues')
plt.suptitle('Top 20 Most Unique Tokens by Sentiment', fontsize=16)
plt.tight_layout(rect=[0, 0, 1, 0.95])
plt.show()
```

```python
tokenizer = Tokenizer(num_words=max_words, oov_token='<OOV>')
tokenizer.fit_on_texts(X_train)
```

```python
def predict_sentiment_with_all_probs(text):
    seq = tokenizer.texts_to_sequences([text])
    pad = pad_sequences(seq, maxlen=max_len, padding='post')
    probs = model.predict(pad).flatten()

    # For binary: show both classes, for multiclass: show all
    if probs.shape[0] == 1:

        # binary, sigmoid output, just "spam"
        prob_dict = {le.classes_[0]: 1 - probs[0], le.classes_[1]: probs[0]}
    else:
        prob_dict = {label: float(prob) for label, prob in zip(le.classes_, probs)}

    # Highest probability
    idx = int(np.argmax(list(prob_dict.values())))
    predicted_label = le.classes_[idx]
    confidence = list(prob_dict.values())[idx]

    return predicted_label, confidence, prob_dict
```

```python
# Garante que o texto está tokenizado
df['tokens'] = df['cleaned_v2'].astype(str).str.split()
```

```python
# Treine o Word2Vec no seu corpus
w2v_model = Word2Vec(
    sentences=df['tokens'],
    vector_size=100,   # tamanho do vetor de cada palavra (ajustável)
    window=5,
    min_count=2,
    sg=1,              # Skip-gram; use sg=0 para CBOW
    workers=4,
    seed=42
)
```

```python
vocab_words = set(w2v_model.wv.index_to_key)
print("spam" in vocab_words)
print(list(vocab_words)[:20])  # Mostra 20 palavras do vocabulário
```

```python
# Mostra as top palavras em textos SPAM
spam_tokens = [word for tokens in df[df['label']=='spam']['tokens'] for word in tokens]
spam_word_counts = Counter(spam_tokens)
print(spam_word_counts.most_common(20))
```

```python
print("Palavras mais semelhantes a 'customer':")
for word, similarity in w2v_model.wv.most_similar('customer', topn=10):
    print(f"{word}: {similarity:.2f}")
```

```python
spam_keywords = [w for w, _ in spam_word_counts.most_common(10) if w in vocab_words]
print(f"Usando palavras: {spam_keywords}")
similar_words = w2v_model.wv.most_similar(positive=spam_keywords, topn=10)
for word, similarity in similar_words:
    print(f"{word}: {similarity:.2f}")
```

```python
# Crie o vetor do texto (Feature Extraction)

def document_vector(tokens, model):
    # Remove palavras fora do vocabulário
    tokens = [word for word in tokens if word in model.wv]
    if len(tokens) == 0:
        return np.zeros(model.vector_size)
    return np.mean(model.wv[tokens], axis=0)

# Cria um array com o vetor de cada texto
X_w2v = np.vstack(df['tokens'].apply(lambda x: document_vector(x, w2v_model)))
```

```python
# Pegue top 30 palavras de spam presentes no vocabulário do Word2Vec
spam_tokens = [word for tokens in df[df['label']=='spam']['tokens'] for word in tokens]
spam_counts = Counter(spam_tokens)
top_spam_words = [w for w, _ in spam_counts.most_common(30) if w in w2v_model.wv]

# Pegue os vetores dessas palavras
word_vectors = [w2v_model.wv[w] for w in top_spam_words]
```

```python
# Converta a lista de vetores para matriz
word_vectors_matrix = np.vstack(word_vectors)

# Reduza para 2 dimensões
pca = PCA(n_components=2, random_state=42)

#
vecs_2d = pca.fit_transform(word_vectors_matrix)

#
pca
```

```python
plt.figure(figsize=(10, 8))
plt.scatter(vecs_2d[:, 0], vecs_2d[:, 1], c='red')

# Adicione os nomes das palavras no gráfico
for i, word in enumerate(top_spam_words):
    plt.text(vecs_2d[i, 0]+0.02, vecs_2d[i, 1]+0.02, word, fontsize=12)

plt.title("PCA (2D) das Palavras Mais Frequentes em SPAM")
plt.xlabel("PCA1")
plt.ylabel("PCA2")
plt.tight_layout()
plt.show()
```
